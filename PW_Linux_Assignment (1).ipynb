### Q2) What is the difference between Hard Link & Soft Link?

**Ans:** In Linux, both hard links and soft links are ways to create references to files,
but they work very differently at the system level.

A hard link is essentially another name for the same file. It points directly to the
same inode, which means both the original file and the hard link share the same data
on disk. When you create a hard link, no new data is created — just an additional
filename that refers to the same content. If the original file is deleted, the data
is still accessible through the hard link because the content exists until all hard
links pointing to it are removed.

Soft link (also known as a symbolic link) is like a shortcut. It does not point to
the data directly but instead stores the path to the original file. This soft link
has its own inode and is treated as a separate file by the system. When you access
a soft link, the system follows the path stored inside it to reach the original file.
If the original file is deleted, the soft link breaks and becomes a non-usable link,
since its target no longer exists. Soft links can point to files across different
filesystems and can also link to directories.

In terms of storage, hard links do not take extra space for the file content because
they use the same inode. Soft links do use a small amount of space, just enough to
store the path to the original file.

**Ans:** kernel is the core part of the Linux OS. It manages CPU and memory usage,
Process scheduling,Device drivers,System calls,File systems. The kernel acts

on your Linux server. How would you identify and terminate this process?

**Ans:** If I suspect that a process is using too much CPU on my Linux system, the
first thing I would do is run the top or htop command. This shows a live list
of all running processes along with their CPU usage. I would look for the process
with the highest %CPU value.

After identifying the PID (Process ID) of that process, I can terminate it using
the kill command like this:
```
kill PID
```

And if it doesn’t stop with the normal kill command, I would forcefully stop it using:

```
kill -9 PID
```
This way, I can control any process that is consuming too many resources and
affecting system performance.
